```python
def simplify_rating(rating):
    if rating in ["TV-Y", "TV-Y7", "TV-Y7-FV", "TV-G", "G"]:
        return "Kids"
    
    elif rating in ["PG", "TV-PG", "PG-13", "TV-14"]:
        return "Teen"
    
    elif rating in ["R", "TV-MA", "NC-17"]:
        return "Adult"
    
    else:
        return "Not Rated"
```

```python
netflix["rating_category"] = netflix["rating"].apply(simplify_rating)
```

```python
netflix["rating_category"].value_counts()
```

```python
netflix["rating"] = netflix["rating"].fillna("Not Rated")
```

```python
netflix["director"] = netflix["director"].fillna("Unknown")
```

```python
netflix["cast"] = netflix["cast"].fillna("Unknown")
```

```python
netflix["country"] = netflix["country"].fillna("Unknown")
```

```python
netflix.drop(columns = ["rating"], inplace = True)
```

```python
netflix["duration_int"] = netflix["duration"].str.extract("(\d+)").astype(int)
```

```python
netflix.drop(columns = "duration", inplace = True)
```

```python
netflix["added_year"] = netflix["date_added"].dt.year
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x = "type", data = netflix)
plt.xlabel("Type")
plt.ylabel("Count")
plt.title("Movies vs TV Shows")
plt.grid(True)
plt.show()
```

```python
top_countries = netflix["country"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values, )
plt.title("Top 10 Countries")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="rating_category", data=netflix)
plt.xlabel("Categories")
plt.ylabel("Count")
plt.title("Audience Category Distribution")
plt.grid(True)
plt.show()
```

```python
movies = netflix[netflix["type"] == "Movie"]

plt.figure(figsize=(8,4))
sns.histplot(movies["duration_int"], bins=30, kde=True)
plt.xlabel("Duration in minutes")
plt.ylabel("Count")
plt.title("Movie Duration Distribution")
plt.grid(True)
plt.show()
```

```python
tv = netflix[netflix["type"] == "TV Show"]

plt.figure(figsize=(8,4))
sns.countplot(x="duration_int", data=tv)
plt.xlabel("Seasons")
plt.ylabel("Count")
plt.title("Number of Seasons Distribution")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
netflix["added_year"].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
yearly_type = netflix.groupby(["added_year", "type"]).size().unstack()
```

```python
plt.figure(figsize=(12,6))

yearly_type.plot(marker="o")

plt.title("Movies vs TV Shows Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```
